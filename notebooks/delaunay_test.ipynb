```python

from scipy.spatial import Voronoi, Delaunay
from dtet import DelaunayTriangulation
import numpy as np

for i in range(5000):
    v = np.random.rand(20, 3)
    ind1 = Delaunay(v).simplices
    d = DelaunayTriangulation()
    d.init_from_points(v)
    ind2 = d.get_cells()
    ind1 = ind1[np.lexsort(ind1.T)]
    ind2 = ind2[np.lexsort(ind2.T)]
    if (len(ind1) != len(ind2)):
        print(len(ind1), len(ind2))
        print(ind2)
```
